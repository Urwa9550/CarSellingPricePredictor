```python
# Function to clean and extract torque and RPM
def extract_torque_rpm(torque_str):
    if not torque_str or pd.isna(torque_str):
        return None, None  # Handle empty or NaN values

    # Step 1: Standardize string: remove spaces and commas, convert to lowercase
    torque_str = torque_str.replace(" ", "").replace(",", "").lower()

    # Step 2: Extract torque value and unit (Nm or kgm)
    torque_match = re.search(r'([\d\.]+)(nm|kgm)?', torque_str)
    if torque_match:
        torque_value = float(torque_match.group(1))
        torque_unit = torque_match.group(2)

        # Convert kgm to Nm (1 kgm ≈ 9.80665 Nm)
        if torque_unit == 'kgm':
            torque_value *= 9.80665
    else:
        torque_value = None

    # Step 3: Extract RPM value(s)
    rpm_match = re.search(r'(\d+)(?:-)?(\d+)?rpm', torque_str)
    if rpm_match:
        # If it's a range like 1750-2750, take the lower bound
        rpm_value = int(rpm_match.group(1))
    else:
        rpm_value = None

    return torque_value, rpm_value

torque_list = cars['torque'].to_list()  # Extracting the 'torque' column as a list

# Apply the function to each item in the torque_list using list comprehension
results = [extract_torque_rpm(torque) for torque in torque_list]

# Convert the result into two separate lists: torque_nm and torque_rpm
torque_nm, torque_rpm = zip(*results)

# Add the two lists as new columns in the original DataFrame
cars['torque_nm'] = torque_nm
cars['torque_rpm'] = torque_rpm

# Display the DataFrame with new columns
#print(cars)
```

```python
# Function to extract mileage value and convert units if necessary
def extract_mileage_kmpl(mileage_str):
    if not mileage_str or pd.isna(mileage_str):
        return None  # Handle empty or NaN values
    
    # Step 1: Extract numeric value
    mileage_match = re.search(r'([\d\.]+)', mileage_str)
    if mileage_match:
        mileage_value = float(mileage_match.group(1))
    else:
        mileage_value = None
    
    # Step 2: Check the unit and convert if necessary
    if 'km/kg' in mileage_str:
        # Conversion factor from km/kg to kmpl (CNG/LPG to petrol equivalent)
        conversion_factor = 1.4  # This is an approximate value, adjust if necessary
        mileage_value *= conversion_factor
    
    return mileage_value

# Outside the function
mileage_list = cars['mileage'].to_list()  # Convert the 'mileage' column to a list
results = [extract_mileage_kmpl(miles) for miles in mileage_list]  # Apply the function to each item in the list

# Adding the results to a new column in the cars DataFrame
cars['mileage_kmpl'] = results
```
